```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('data/housing.csv')
sample = df.sample(n=1000, random_state=42)
df.head()
```

```python
alt.Chart(sample).mark_point().encode(
    x='longitude',
    y='latitude',
    color='median_house_value'
).properties(
    width=600,
    height=600
).interactive()
```
